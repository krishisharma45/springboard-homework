## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
#Problem 1: List the 10 countries with the most projects
import pandas as pd
import json
from pandas.io.json import json_normalize

data = 'data/world_bank_projects.json'
projects = pd.read_json(data)
cols = projects[['_id', 'countryshortname', 'countryname', 'countrycode', 'theme_namecode']]
cols.groupby('countryname').count().sort_values(by='countryshortname', ascending=False)[:10]
```

```python
#Problem 2: Find the top 10 major project themes
json.load(open(data))[0]['mjtheme_namecode']
msgs = pd.io.json.json_normalize(json.load(open(data)), record_path="mjtheme_namecode")
msgs.groupby('name').count().sort_values(by='code', ascending=False)[1:11]
```

```python
##I tried to use fillna, but as the analysis shows, for some reason, the empty records in the name column
##are not considered null because the dtypes are objects. I have tried to convert the names column to strings,
##but I am stuck.
msgs = pd.io.json.json_normalize(json.load(open(data)), record_path="mjtheme_namecode")
print(len(msgs))
print(msgs['name'].isnull().sum())
print(msgs.dtypes)
msgs.code = msgs.code.astype(float)
msgs.name = msgs.name.astype(str).str.replace(',','')
msgs.dtypes
```

```python
##I have demonstrated my code for how I would use fillna if the missing records in the name column were actually empty.
#Problem 3: Create a dataframe that fills in the empty names for mjtheme_namecodes
values = pd.Series(msgs.name.values, index=msgs.code).to_dict()
values = {k: v for k, v in values.items() if v is not ""}
msgs["name"] = msgs["name"].fillna(value = values, inplace=True)
```

```python
##This is the original solution. I tried to replace the empty strings with NaN to use fillna, but that did not work.
#Problem 3: Create a dataframe that fills in the empty names for mjtheme_namecodes
msgs = pd.io.json.json_normalize(json.load(open(data)), record_path="mjtheme_namecode")
msgs1 = msgs[msgs.name == ""].reset_index()
msgs2 = msgs[msgs.name != ""].reset_index()
msgs2 = msgs2.rename(columns={'code':'code_x', 'name':'name_y'})
df = msgs1.merge(msgs2, how='left', right_on='code_x', left_on='code')
df = df.drop_duplicates(keep='first', inplace=False, subset=["index_x", "code"])
df = df[['code', 'name_y']]
df.rename(columns={'code':'code', 'name_y':'name'})
```
